```python
import numpy as np
import pandas as pd

cal = pd.read_csv("calories.csv")
cal
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
restaurant_names = ["Subway", "Sonic", "Arbys", "Taco Bell", "Chick Fil-A", "Dairy Queen", "Mcdonalds"]
total_calories = [3930, 3220, 2640, 2640, 2530, 2370, 2280]

# Generate a color map with a gradient from blue to ochre
colors = [plt.cm.viridis(i) for i in np.linspace(0.1, 0.9, len(restaurant_names))]

# Create a horizontal bar graph
plt.figure(figsize=(10, 6))
bars = plt.barh(restaurant_names, total_calories, color=colors)

# Add values inside the bars
for bar in bars:
    plt.text(bar.get_width() - 200, bar.get_y() + bar.get_height() / 2, 
             f'{bar.get_width()}', 
             va='center', ha='center', color='white', fontweight='bold')

# Add title and labels
plt.title('Average Calories by Restaurant')
plt.xlabel('Average Calories Kcal in 100g')
plt.ylabel('Restaurant')

# Save the figure as a PNG file
plt.savefig("Average calories.png", format='png', bbox_inches='tight')

# Display the horizontal bar graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data (excluding "Not Fast Food")
restaurant_names = ["Chick Fil-A", "Taco Bell", "Sonic", "Arbys", "Dairy Queen", "Subway", "Mcdonalds"]
avg_sugar = [3.400000, 4.600000, 5.500000, 6.000000, 6.000000, 6.615385, 7.333333]

# Generate a color map with a gradient from blue to ochre
colors = [plt.cm.viridis(i) for i in np.linspace(0.1, 0.9, len(restaurant_names))]

# Create a horizontal bar graph
plt.figure(figsize=(12, 7))
bars = plt.barh(restaurant_names, avg_sugar, color=colors)

# Add values inside the bars
for bar in bars:
    plt.text(bar.get_width() - 0.2, bar.get_y() + bar.get_height() / 2, 
             f'{bar.get_width():.2f}', 
             va='center', ha='center', color='white', fontweight='bold')

# Add title and labels
plt.title('Average Sugar Content by Restaurant')
plt.xlabel('Average Sugar Content (g) in 100g')
plt.ylabel('Restaurant')

# Save the figure as a PNG file
plt.savefig("fast_food_sugar_content.png", format='png', bbox_inches='tight')

# Display the horizontal bar graph
plt.show()
```
